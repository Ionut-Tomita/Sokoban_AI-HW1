# Solving a map 

```python
# change the map name to test different maps

map = Map.from_yaml('tests/large_map2.yaml')
```

# Visualizing a map 

```python
print(map)
```

```python
map.plot_map()
```

```python
# To save the state of a map to disk:
map.save_map('images', 'map.png')
```

# Run algorithms

```python
# Rulează algoritmul Beam Search
print("Running Beam Search...")
start_time = time.time()
solution, total_steps = beam_search(
    map_obj=map,
    beam_width=20,  # Ajustează lățimea fasciculului
    max_iterations=10000,
)
end_time = time.time()

# Afișează rezultatele
if solution:
    print(f"Solution found in {end_time - start_time:.2f} seconds!")
    print(f"Number of steps: {len(solution)}")
    save_images(solution, 'images/beam_map')
    print(f"Total steps explored by the algorithm: {total_steps}")

    print("Solution steps saved as images.")

    create_gif('images/beam_map', 'beam_map', 'images')
    print("GIF created: images/beam_map.gif")
else:
    print("No solution found.")

```

```python
# Rulează algoritmul Idastar
print("Running IDA* Search...")
start_time = time.time()
solution, total_steps = ida_star(
    map_obj=map,
    max_iterations=10000,
)
end_time = time.time()
# Afișează rezultatele
if solution:
    print(f"Solution found in {end_time - start_time:.2f} seconds!")
    print(f"Number of steps: {len(solution)}")
    print(f"Total steps explored by the algorithm: {total_steps}")
    save_images(solution, 'images/idastar_map')
    print("Solution steps saved as images.")

    create_gif('images/idastar_map', 'idastar_map', 'images')
    print("GIF created: images/idastar_map.gif")
else:
    print("No solution found.")

```
